```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
import pickle

# Load the merged dataset
merged_data = pd.read_csv('merged_data.csv')

# Check if the necessary columns are present and calculate BMI if not already present
if 'BMI' not in merged_data.columns:
    if 'Height' in merged_data.columns and 'Weight' in merged_data.columns:
        # Calculate BMI and add it as a column
        merged_data['BMI'] = merged_data['Weight'] / ((merged_data['Height'] / 100) ** 2)
        print("BMI column added successfully.")
    else:
        print("Columns 'Height' and/or 'Weight' are missing in the dataset.")

# Encode 'Gender' column
label_encoder = LabelEncoder()
merged_data['Gender'] = label_encoder.fit_transform(merged_data['Gender'])  # male = 1, female = 0

# Select relevant features and labels
features = merged_data[['Age', 'BMI', 'Duration', 'Heart_Rate', 'Body_Temp', 'Gender']]
labels = merged_data['Calories']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Initialize and train the model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Save the trained model to a file
with open('fitness_model.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)

print('Model trained and saved successfully as fitness_model.pkl')
```

```python
%%writefile app.py
import streamlit as st
import pickle
import numpy as np

# Load the trained model
model = pickle.load(open('fitness_model.pkl', 'rb'))

# Title of the app
st.title('Personal Fitness Tracker')

# Sidebar for user inputs
st.sidebar.header('User Input Parameters')

def user_input_features():
    Gender = st.sidebar.selectbox('Gender', ('Male', 'Female'))
    Age = st.sidebar.slider('Age', 10, 80, 25)
    Height = st.sidebar.slider('Height (in cm)', 100, 250, 170)
    Weight = st.sidebar.slider('Weight (in kg)', 30, 150, 70)
    Duration = st.sidebar.slider('Duration (in minutes)', 0, 300, 30)
    Heart_Rate = st.sidebar.slider('Heart Rate', 60, 200, 80)
    Body_Temp = st.sidebar.slider('Body Temperature (in °C)', 35.0, 42.0, 37.0)

    # Calculate BMI
    BMI = Weight / ((Height / 100) ** 2)

    # Encode gender
    Gender_male = 1 if Gender == 'Male' else 0

    # Prepare the features as a numpy array
    features = np.array([[Age, BMI, Duration, Heart_Rate, Body_Temp, Gender_male]])

    return features

# Get user input features
input_data = user_input_features()

# Make prediction
if st.sidebar.button('Predict Calories Burned'):
    prediction = model.predict(input_data)[0]
    st.write(f'Estimated Calories Burned: {prediction:.2f} kcal')
```
